### Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation of 5 using Python. Interpret the results.

```python
import scipy.stats as stats

sample_mean = 50
sample_stddev = 5
confidence_level = 0.95 
sample_size = 20

standard_error = sample_stddev / (sample_size ** 0.5)

margin_of_error = stats.norm.ppf(1 - (1 - confidence_level) / 2) * standard_error

lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

### Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag matches the expected distribution of 20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20% brown. Use Python to perform the test with a significance level of 0.05.

```python
import numpy as np
from scipy.stats import chisquare

# let frequencies of colors in the M&Ms bags
observed_frequencies = np.array([50, 45, 55, 15, 18, 42])  

expected_distribution = np.array([0.20, 0.20, 0.20, 0.10, 0.10, 0.20])

# expected frequencies based on the expected distribution and the total sample size
total_sample_size = np.sum(observed_frequencies)
expected_frequencies = expected_distribution * total_sample_size

# chi-square goodness of fit test
chi_square_stat, p_value = chisquare(observed_frequencies, f_exp=expected_frequencies)

alpha = 0.05

print(f"Chi-Square Statistic: {chi_square_stat:.2f}")
print(f"P-value: {p_value:.5f}")

if p_value < alpha:
    print("Reject the null hypothesis: The observed distribution is not consistent with the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The observed distribution is consistent with the expected distribution.")
```

###  Use Python to calculate the chi-square statistic and p-value for a contingency table with the following data:



```python
import numpy as np
from scipy.stats import chi2_contingency

observed_data = np.array([[20, 10, 15], [15, 25, 20]])

chi2_stat, p_value, dof, expected = chi2_contingency(observed_data)

alpha = 0.05

print(f"Chi-Square Statistic: {chi2_stat:.2f}")
print(f"P-value: {p_value:.5f}")

if p_value < alpha:
    print("Reject the null hypothesis: The groups and outcomes are not independent.")
else:
    print("Fail to reject the null hypothesis: The groups and outcomes are independent.")
```

**Conclusion** The calculated chi-square statistic is 5.83. This value is a measure of how much the observed frequencies deviate from what would be expected if the groups and outcomes were independent. In essence, it quantifies the difference between the observed and expected frequencies. The calculated p-value associated with the chi-square statistic is approximately 0.05411. This p-value represents the probability of observing the data if the groups and outcomes are truly independent (null hypothesis is true). Since the p-value (0.05411) is greater than the significance level (alpha) of 0.05, we fail to reject the null hypothesis. This means that the evidence in the data is not strong enough to suggest that there is a significant relationship between the groups and outcomes.

### A study of the prevalence of smoking in a population of 500 individuals found that 60 individuals smoked. Use Python to calculate the 95% confidence interval for the true proportion of individuals in the population who smoke.

```python
import scipy.stats as stats
import numpy as np

total_population = 500

# proportion of individuals who smoke
sample_proportion = 60 / total_population
# sample_proportion = 0.12

confidence_level = 0.95  

# Wilson Score Interval
z = stats.norm.ppf(1 - (1 - confidence_level) / 2)
denominator = 1 + z**2 / total_population
center_adjusted_proportion = sample_proportion + z**2 / (2 * total_population)
adjusted_standard_deviation = np.sqrt((sample_proportion * (1 - sample_proportion) + z**2 / (4 * total_population)) / total_population)
lower_bound = (center_adjusted_proportion - z * adjusted_standard_deviation) / denominator
upper_bound = (center_adjusted_proportion + z * adjusted_standard_deviation) / denominator

lower_bound_percentage = lower_bound * 100
upper_bound_percentage = upper_bound * 100

print(f"95% Confidence Interval for Smoking Proportion: ({lower_bound_percentage:.2f}%, {upper_bound_percentage:.2f}%)")
```

### Calculate the 90% confidence interval for a sample of data with a mean of 75 and a standard deviation of 12 using Python. Interpret the results.

```python
import scipy.stats as stats

sample_mean = 75
sample_stddev = 12
sample_size = 100    # let
confidence_level = 0.90  

standard_error = sample_stddev / (sample_size ** 0.5)

degrees_of_freedom = sample_size - 1

t_score = stats.t.ppf(1 - (1 - confidence_level) / 2, df=degrees_of_freedom)

margin_of_error = t_score * standard_error

lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print(f"90% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

The calculated 90% confidence interval is (73.01, 76.99). This means that we are 90% confident that the true population mean lies within the interval (73.01, 76.99) based on the sample data.
In other words, if we were to take many different samples from the same population and calculate the confidence interval for each sample, we expect that about 90% of those intervals would contain the true population mean.

###  Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the area corresponding to a chi-square statistic of 15.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

df = 10
x = np.linspace(1, 50, 500)  

chi2_pdf = stats.chi2.pdf(x, df)

plt.figure(figsize=(10, 5))
plt.plot(x, chi2_pdf, label=f"Chi-Square with {df} Degrees of Freedom")

# shade the area corresponding to chi-square statistic of 15
x_shade = np.linspace(15, 30, 500)
y_shade = stats.chi2.pdf(x_shade, df)
plt.fill_between(x_shade, y_shade, color='gray', alpha=0.5, label="Shaded Area (statistic = 15)")

plt.xlabel("Chi-Square Statistic")
plt.ylabel("Probability Density")
plt.title("Chi-Square Distribution")
plt.legend()
plt.grid()
plt.show()
```

### A random sample of 1000 people was asked if they preferred Coke or Pepsi. Of the sample, 520 preferred Coke. Calculate a 99% confidence interval for the true proportion of people in the population who prefer Coke.

```python
import scipy.stats as stats
import numpy as np

sample_size = 1000
sample_proportion = 520 / sample_size
confidence_level = 0.99 

standard_error = np.sqrt((sample_proportion * (1 - sample_proportion)) / sample_size)

z_score = stats.norm.ppf(1 - (1 - confidence_level) / 2)

margin_of_error = z_score * standard_error

lower_bound = sample_proportion - margin_of_error
upper_bound = sample_proportion + margin_of_error

lower_bound_percentage = lower_bound * 100
upper_bound_percentage = upper_bound * 100

print(f"99% Confidence Interval for Proportion Preferring Coke: ({lower_bound_percentage:.2f}%, {upper_bound_percentage:.2f}%)")
```

###  A researcher hypothesizes that a coin is biased towards tails. They flip the coin 100 times and observe 45 tails. Conduct a chi-square goodness of fit test to determine if the observed frequencies match the expected frequencies of a fair coin. Use a significance level of 0.05.

```python
import numpy as np
from scipy.stats import chisquare

observed_frequencies = np.array([55, 45])  # observed tails, observed head
expected_frequencies = np.array([50, 50])  # Expected tails, expected heads (fair coin)

chi_square_stat, p_value = chisquare(observed_frequencies, f_exp=expected_frequencies)

alpha = 0.05

print(f"Chi-Square Statistic: {chi_square_stat:.2f}")
print(f"P-value: {p_value:.5f}")

if p_value < alpha:
    print("Reject the null hypothesis: The coin is biased towards tails.")
else:
    print("Fail to reject the null hypothesis: The coin is not significantly biased.")
```

###  A random sample of 30 people was selected from a population with an unknown mean and standard deviation. The sample mean was found to be 72 and the sample standard deviation was found to be 10. Conduct a hypothesis test to determine if the population mean is significantly different from 70. Use a significance level of 0.05.

```python
import scipy.stats as stats

sample_mean = 72
sample_stddev = 10
sample_size = 30
population_mean_hypothesized = 70  # Hypothesized population mean
confidence_level = 0.95  

standard_error = sample_stddev / (sample_size ** 0.5)

t_score = (sample_mean - population_mean_hypothesized) / standard_error

degrees_of_freedom = sample_size - 1

t_critical = stats.t.ppf(1 - (1 - confidence_level) / 2, df=degrees_of_freedom)

p_value = 2 * (1 - stats.t.cdf(abs(t_score), df=degrees_of_freedom))

print(f"T-Score: {t_score:.2f}")
print(f"Critical T-Value: {t_critical:.2f}")
print(f"P-value: {p_value:.5f}")

if abs(t_score) > t_critical or p_value < 0.05:
    print("Reject the null hypothesis: The population mean is significantly different from 70.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in the population mean.")
```
