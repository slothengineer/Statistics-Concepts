### What is the Probability density function?

The Probability Density Function (PDF) is a fundamental concept in probability theory and statistics. It is used to describe the likelihood of a continuous random variable taking on a specific value within a certain range. In other words, the PDF provides a way to model the distribution of a continuous random variable.

For a continuous random variable (X), the PDF is denoted as f(x), where f(x) gives the probability of (X) taking a value in the infinitesimally small interval ([x, x + ▲x]), where (▲x) is a very small increment. Mathematically, the PDF is defined as the derivative of the cumulative distribution function (CDF) with respect to (x):

               f(x) = d/dx*[F(x)

Here, F(x) is the cumulative distribution function, which gives the probability that (X) is less than or equal to (x). The PDF f(x) is non-negative for all values of (x), and the area under the PDF curve over the entire range of (X) is equal to 1, since the total probability of all possible outcomes is 1.

Properties of the PDF:
1. The PDF is non-negative: f(x)>=0 for all (x).
2. The integral of the PDF over its entire range is equal to 1: ∫ (−∞,∞) f(x).dx = 1.
3. The probability that X falls within a certain interval [a, b] is given by the integral of the PDF over that interval:
   ∫ (a,b) f(x).dx represents the probability that (a <= X <= b).

Different types of continuous probability distributions, such as the normal distribution, exponential distribution, and uniform distribution, have their own specific PDFs that describe how the values are distributed. The shape of the PDF provides information about the likelihood of different values occurring in the data.

### What are the types of Probability distribution?

Probability distributions are mathematical functions that describe how the probabilities of different outcomes are distributed for a given random variable.

1. **Discrete Probability Distributions:**

   - **Bernoulli Distribution:** Models a single binary (yes/no) outcome with a fixed probability of success.
   - **Binomial Distribution:** Represents the number of successes in a fixed number of independent Bernoulli trials.
   - **Poisson Distribution:** Describes the number of events that occur in a fixed interval of time or space, given an average rate.

2. **Continuous Probability Distributions:**

   - **Normal Distribution (Gaussian Distribution):** Characterized by a symmetric bell-shaped curve. Many real-world phenomena follow this distribution due to the central limit theorem.
   - **Exponential Distribution:** Models the time between events in a Poisson process, often used in reliability and queueing theories.
   - **Uniform Distribution:** All outcomes in a given range are equally likely.
   - **Gamma Distribution:** Generalizes the exponential distribution and is used in areas such as queuing theory, reliability analysis, and physics.
   - **Beta Distribution:** Represents the distribution of probabilities for events that are bounded by a minimum and maximum value.
   - **Chi-Square Distribution:** Arises in statistical hypothesis testing and is related to the sum of squared standard normal random variables.
   - **Student's t-Distribution:** Used in hypothesis testing when the sample size is small and the population standard deviation is unknown.
   - **F-Distribution:** Often used in the analysis of variance (ANOVA) and regression analysis.

3. **Multivariate Probability Distributions:**

   - **Multinomial Distribution:** Generalization of the binomial distribution to more than two outcomes.
   - **Multivariate Normal Distribution:** Generalization of the normal distribution to multiple variables.

### Write a Python function to calculate the probability density function of a normal distribution withgiven mean and standard deviation at a given point.



```python
import math

def normal_pdf(x, mean, std_dev):
    exponent = -((x - mean) ** 2) / (2 * std_dev ** 2)
    pdf_value = (1 / (std_dev * math.sqrt(2 * math.pi))) * math.exp(exponent)
    return pdf_value

mean = 0  
std_dev = 1  
x_value = 1.5 

pdf_at_x = normal_pdf(x_value, mean, std_dev)
print(f"PDF at x = {x_value}: {pdf_at_x}")
```

### What are the properties of Binomial distribution? Give two examples of events where binomial distribution can be applied.

The binomial distribution is a discrete probability distribution that models the number of successes in a fixed number of independent Bernoulli trials, where each trial has two possible outcomes: success or failure. The properties of the binomial distribution include:

1. **Fixed Number of Trials:** The binomial distribution is applicable when there is a fixed number n of independent trials.

2. **Independent Trials:** Each trial is independent of the others, meaning the outcome of one trial doesn't affect the outcomes of other trials.

3. **Two Outcomes:** Each trial has two possible outcomes: success (usually denoted as S) or failure (usually denoted as F).

4. **Constant Probability:** The probability of success (p) and the probability of failure (q = 1 - p) remain constant from trial to trial.

5. **Count of Successes:** The binomial random variable (X) represents the number of successes in (n) trials.

6. **Probability Mass Function (PMF):** The probability mass function of the binomial distribution gives the probability of observing exactly (k) successes in (n) trials and is given by the formula:
   
   P(X = k) = binom({n}{k}).p^k q^{n-k}
   
   Where (binom{n}{k}) is the binomial coefficient, equal to (n!/k!*(n-k)!).

7. **Mean and Variance:** The mean (µ) of a binomial distribution is ( n.p), and the variance (σ^2) is ( n . p . q ).

Examples of events where the binomial distribution can be applied:

1. **Coin Flips:** Consider the scenario of flipping a fair coin multiple times. Each flip has two possible outcomes: heads (success) or tails (failure). The binomial distribution can be used to model the number of times heads appears in a fixed number of coin flips.

2. **Quality Control:** In manufacturing, a company might perform quality control tests on a sample of products. Each product either passes (success) or fails (failure) the test. The binomial distribution can be used to model the number of products that pass the test in a fixed number of sampled products.

In both of these examples, we have a fixed number of trials (coin flips or quality control tests), each trial has two possible outcomes (success or failure), and the trials are independent of each other. These properties make the binomial distribution a suitable model for such scenarios.

### Generate a random sample of size 1000 from a binomial distribution with probability of success 0.4 and plot a histogram of the results using matplotlib.

### Write a Python function to calculate the cumulative distribution function of a Poisson distribution with given mean at a given point.



```python
import math

def poisson_cdf(x, mean):
    cdf_value = 0
    for k in range(x + 1):
        cdf_value += (math.exp(-mean) * mean**k) / math.factorial(k)
    return cdf_value

mean = 3.5  
x_value = 2

cdf_at_x = poisson_cdf(x_value, mean)
print(f"CDF at x = {x_value}: {cdf_at_x}")
```

### How Binomial distribution different from Poisson distribution?

The binomial distribution and the Poisson distribution are both probability distributions used to model the number of events or occurrences, but they have different characteristics and are suitable for different types of scenarios. Here's how they differ:

**1. Nature of Events:**
   - **Binomial Distribution:** The binomial distribution models the number of successes in a fixed number of independent trials, where each trial has two possible outcomes: success or failure. The probability of success remains constant across trials.
   - **Poisson Distribution:** The Poisson distribution models the number of events that occur in a fixed interval of time or space, given an average rate. The events are assumed to occur randomly and independently.

**2. Number of Trials/Events:**
   - **Binomial Distribution:** Requires a fixed number of trials or events (n).
   - **Poisson Distribution:** Does not require a fixed number of events; it's concerned with the number of events in a fixed interval.

**3. Probability of Success/Event:**
   - **Binomial Distribution:** Requires a constant probability of success (p) in each trial.
   - **Poisson Distribution:** Does not involve probabilities of success; it's determined by the average rate (ƛ).

**4. Mean and Variance:**
   - **Binomial Distribution:** The mean (µ) is (n.p) and the variance (σ^2) is (n.p.q), where (q = 1 - p).
   - **Poisson Distribution:** The mean (µ) and the variance (σ^2) are both equal to (ƛ).

**5. Applications:**
   - **Binomial Distribution:** Used when dealing with situations where you have a fixed number of trials and each trial has two possible outcomes, such as coin flips, pass/fail tests, etc.
   - **Poisson Distribution:** Used when dealing with events that occur randomly and independently over a fixed interval of time or space, such as the number of phone calls received in a given hour, the number of accidents in a day, etc.

**6. Shape:**
   - **Binomial Distribution:** The shape is symmetric if (n) is large and (P) is not too close to 0 or 1.
   - **Poisson Distribution:** The shape is skewed to the right and becomes more symmetric as (ƛ) increases.

### Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the sample mean and variance.

```python
import numpy as np

sample_size = 1000
mean = 5

random_sample = np.random.poisson(mean, size=sample_size)

sample_mean = np.mean(random_sample)
sample_variance = np.var(random_sample)

print(f"Sample Mean: {sample_mean}")
print(f"Sample Variance: {sample_variance}")
```

### How mean and variance are related in Binomial distribution and Poisson distribution?

The mean (µ) and variance (σ^2) in both the binomial distribution and the Poisson distribution are related to the parameters of the distributions, and they have specific formulas that are derived from those parameters. Here's how the mean and variance are related in each distribution:

**Binomial Distribution:**
For a binomial distribution with parameters n (number of trials) and p (probability of success in each trial):

- Mean (µ) = (n . p)
- Variance (σ^2) = (n . p . (1 - p))

The mean of the binomial distribution is the product of the number of trials 'n' and the probability of success in each trial 'p'. The variance is the product of n, p, and (1 - p).

The relationship between mean and variance in the binomial distribution is influenced by the probability of success p. When p is small or large, the variance tends to be larger because the outcomes are skewed toward one end of the distribution. When p is around 0.5, the distribution is symmetric and the variance is relatively smaller.

**Poisson Distribution:**
For a Poisson distribution with parameter ƛ (average rate of events in a fixed interval):

- Mean (µ) = (ƛ)
- Variance (σ^2) = (ƛ)

In the Poisson distribution, both the mean and variance are equal to the parameter ƛ, which represents the average rate of events occurring in the interval.

### In normal distribution with respect to mean position, where does the least frequent data appear?

In a normal distribution (also known as the Gaussian distribution), which is symmetric and bell-shaped, the least frequent data points are located in the tails of the distribution, farthest from the mean. The tails of the distribution contain the extreme values that are less likely to occur compared to values closer to the mean.

In a standard normal distribution (a specific type of normal distribution with a mean of 0 and a standard deviation of 1), about 5% of the data lies in each tail beyond approximately 1.96 standard deviations from the mean. This area corresponds to the 95% confidence interval in statistics.

In a general normal distribution with a mean (µ) and standard deviation (σ), the least frequent data points will be those that are located at the tails beyond a few standard deviations from the mean, where the probability density decreases significantly. The farther you move away from the mean, the lower the probability of observing data points in those regions.

It's important to note that in a normal distribution, there are technically no "least frequent" data points in an absolute sense. The distribution is continuous and unbounded, which means that every data point has a non-zero probability of occurring. However, data points in the tails are less likely to be observed compared to those near the mean, making them relatively less frequent.
